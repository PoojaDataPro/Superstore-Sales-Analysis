```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
df = pd.read_csv(r"C:\Users\Hp\Desktop\Retail Sales Analysis\Data\Superstore_sales\Superstore_sales.csv" )
print(df.head())
```

```python
segment_sales=df.groupby('Segment')['Sales'].sum().sort_values(ascending=True)
print(segment_sales)
```

```python
top_products=df.groupby('Product Name')['Sales'].sum().sort_values(ascending=True).head(10)
print(top_products)
```

```python
shipping_sales = df.groupby('Ship Mode')['Sales'].sum().sort_values(ascending=False)
print(shipping_sales)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Year'] = df['Order Date'].dt.year
yearly_sales = df.groupby('Year')['Sales'].sum()
print(yearly_sales)
```

```python
df['Month'] = df['Order Date'].dt.month
monthly_sales = df.groupby('Month')['Sales'].sum()
print(monthly_sales)
```
